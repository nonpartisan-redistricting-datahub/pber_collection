```python
count = 0
for change in updates_list:
    
    from_list = change["From"]
    to_list = change["To"]
    block_list = change["Blocks"]
    
    print(change)
    print(count)
    
    

    ax = ga_pber[ga_pber["UNIQUE_ID"].isin(to_list)].plot(color = "red", alpha = .5)
    ga_pber[ga_pber["UNIQUE_ID"].isin(from_list)].plot(ax = ax, color = "green", alpha = .5)
    ga_blocks[ga_blocks["GEOID20"].isin(block_list)].boundary.plot(ax = ax, color = "green")
    ax.set_title(count)
    count += 1
    added_geoms = gp.overlay(ga_pber[ga_pber["UNIQUE_ID"].isin(to_list)], ga_blocks[ga_blocks["GEOID20"].isin(block_list)], how = "union", keep_geom_type = True)
    subtracted_geoms = gp.overlay(ga_pber[ga_pber["UNIQUE_ID"].isin(from_list)], added_geoms, how = "difference", keep_geom_type = True)
    
    added_geoms = added_geoms.dissolve()
    subtracted_geoms = subtracted_geoms.dissolve("UNIQUE_ID")
    subtracted_geoms.reset_index(inplace = True, drop = False)
    
    added_geoms = added_geoms[list(ga_pber.columns)]
    subtracted_geoms = subtracted_geoms[list(ga_pber.columns)]
    
    added_geoms.boundary.plot()
    subtracted_geoms.boundary.plot()
    print(ga_pber.shape)
    ga_pber = ga_pber[~ga_pber["UNIQUE_ID"].isin(from_list + to_list)]

    ga_pber = gp.GeoDataFrame(pd.concat([ga_pber, added_geoms, subtracted_geoms]), crs = ga_pber.crs)
    print(ga_pber.shape)
    ga_pber.reset_index(drop = True, inplace = True)
    
    
```
